```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

raw_data = pd.read_csv('/content/drive/MyDrive/raw_student_data.csv')
cleaned_data = raw_data.copy()

directories = ['data_raw', 'data_cleaned', 'analysis_scripts', 'results', 'report', 'figures', 'code']
for directory in directories:
    if not os.path.exists(directory):
        os.makedirs(directory)

cleaned_data.to_csv('data_cleaned/cleaned_student_data.csv', index=False)



```

```python
# Visualization 1: Box plot of math scores by education level
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='math score', data=cleaned_data)
plt.title('Math Scores by Education Level')
plt.savefig('results/visualization_2.png')
plt.show()

# Visualization 2: Pair plot for numerical variables
plt.figure(figsize=(12, 8))
sns.pairplot(cleaned_data, hue='gender')
plt.title('Pair Plot for Numerical Variables')
plt.savefig('results/visualization_3.png')
plt.show()

# Visualization 3: Distribution of reading scores
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['reading score'], bins=30, kde=True)
plt.title('Distribution of Reading Scores')
plt.savefig('results/visualization_4.png')
plt.show()

# Visualization 4: Correlation matrix heatmap
correlation_matrix = cleaned_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.savefig('results/visualization_5.png')
plt.show()
```
